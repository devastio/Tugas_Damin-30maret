```python
import numpy as np
import matplotlib.pyplot as plt

data = [[23,43],[50,5],[18,2],[38,23],[42,58],[78,56],[12,45],[11,25],[34,35],[65,65],[9,23],[5,45],[8,23],[22,76],[11,66],[23,32],[67,9],[43,54],[67,34]]
dataset = np.array(data)

import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

dendogram = sch.dendrogram(sch.linkage(dataset,method = 'ward'))
```

```python
for i in range (0, len(dataset)):
    plt.scatter(dataset[i][0], dataset[i][1], color = 'b')
```

```python
plt.scatter(dataset[:,0], dataset[:,1], color = 'b')
```

```python
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'ward')
```

```python
y_hc = hc.fit_predict(dataset)
```

```python
plt.scatter(dataset[y_hc ==0,0], dataset[y_hc==0,1],color = 'c')
plt.scatter(dataset[y_hc ==1,0], dataset[y_hc==1,1],color = 'b')
plt.scatter(dataset[y_hc ==2,0], dataset[y_hc==2,1],color = 'r')
plt.scatter(dataset[y_hc ==3,0], dataset[y_hc==3,1],color = 'y')
```
